```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```
df=pd.read_csv("students_placement_data.csv")
df.head()
```

1) line graphs

 line graph between ssc and intermediate/diploma percentage.

2) Scatter Plots

 Scatter plot between SSC and inter/diploma percentage.

```
x=df["SSC Percentage"]
y=df["inter_Diploma_percentage"]
plt.scatter(x,y)
plt.xlabel("SSC marks")
plt.ylabel("inter/diploma marks")
plt.title("Scatter plot between SSC and inter/diploma marks")
```

Do you find any correlation by just visual inspection? If yes, then what kind of correlation you can see? (+ve, -ve or No)


yes, i found correlation by just visual inspection (postively correlated both variables)

3)Histograms

A histogram for B.Tech percentage.

```
bins=np.linspace(0,100,10)
print(bins)
plt.hist(df["B.Tech_percentage"],bins,width=9)
plt.xlabel("B.Tech Marks")
```

How many people have scores between 60-80% in B.tech. Tell your approximate answer by visual inspection?

Nearly 65 students scored between 60-80% in B.Tech.

4)Subplots


Draw a scatter plot between
·   Intermediate and SSC
·   Intermediate and B.Tech
in the same graph.


```
fig,ax=plt.subplots()
ax.scatter(df["inter_Diploma_percentage"],df["SSC Percentage"],label="Ïnter-SSC")
ax.scatter(df["inter_Diploma_percentage"],df["B.Tech_percentage"],label="Inter-B.Tech")
ax.set_xlabel("Intermediate Marks")
ax.legend(loc=4)
```

What's your observation?

The students who got good marks in intermediate also got  good marks in ssc.BUt ,those students scored less in B.Tech.
